# 简单线性回归

黄荣贵（复旦大学社会学系）

```r
library(car)
```

```r
head(Prestige)
```

```r
srm = lm(prestige ~ income, data=Prestige)

print(srm)
```

```r
summary(srm)
```

```r
coef(srm)
```

下面画图命令仅适用于简单线性回归

```r
plot(x=Prestige$income, y=Prestige$prestige)

abline(reg=srm, col="blue")
```

```r
vcov(srm)
```

```r
sqrt(diag(vcov(srm)))
```

```r
predict(srm)
```

```r
residuals(srm)
```

```r
summary(residuals(srm))
```

全标准化回归系数

```r
summary(lm(scale(prestige) ~ scale(income), data=Prestige))
```

对x标准化的标准化回归系数

```r
summary(lm(prestige ~ scale(income), data=Prestige))
```

对自变量进行对数变换后引入模型

```r
summary(lm(prestige ~ log(income), data=Prestige))
```
