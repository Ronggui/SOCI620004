```r
# generate a random variable
set.seed(1000)
plot.ecdf(rnorm(80),verticals=T, cex=0.5, main="ECDF of a normal distribution sample")
#不光滑

curve(pnorm, add=T, col="red")
```

```r
# qq plot
set.seed(100)
x = rnorm(100, mean=50, sd=10)

qqPlot(x) # car
```

```r
# qqplots of random variables governed by different distributions
set.seed(100)
qqPlot(rchisq(100, df=2))
```

```r
# shape of distributions
x = seq(-4,4,by=0.01)
y1 = dnorm(x)
y2 = dchisq(x, df=2)

plot(y2~x, type="l")
lines(x, y1, col="red")
legend(2, 0.45, c("chisq","norm"),
       col=c("black","red"), lty=1)
```

```r
set.seed(20)

qqPlot(rnorm(500, mean=30, sd=10),
       distribution="t", df=3)
```

```r
set.seed(17)
y = rt(100, df=2)
qqPlot(y)
```

```r
x = seq(-4,4,by=0.01)
y1 = dnorm(x)
y2 = dt(x, df=2)

plot(y1~x, type="l")

lines(x, y2, col="red")

legend(2, 0.38, c("t","norm"), col=c("red","black"), lty=1)
```
