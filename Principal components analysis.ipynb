使用特征值分解方法进行主成分分析

```r
# 主成分分析，使用相关矩阵（而不是方差协方差矩阵）

ans = princomp(~ Murder + Assault + Rape, data = USArrests, cor = TRUE)

print(ans)
```

```r
summary(ans)
```

```r
loadings(ans)
```

```r
plot(ans)
```

```r
# variable-obversation relationship
biplot(ans)
```

```r
# combine principal scores with orginal dataset
newdata = cbind(USArrests,ans$scores)

newdata
```

使用奇异值分解方法进行主成分分析

将scale和center参数设置为TRUE，表示对数据进行标准化后再进行主成分分析。

```r
ans2 = prcomp(~ Murder + Assault + Rape, data = USArrests, scale = TRUE, center=TRUE)

ans2
```

注意负荷的符号与princomp所得结果相反。这并不是错误。

```r
summary(ans2)
```

```r
plot(ans2)
```

```r
biplot(ans2)
```

```r
predict(ans2)
```
